```python
nilai = []

for i in range(10):
    n = float(input(f"Masukkan nilai ke-{i+1}: "))
    nilai.append(n)

# Menampilkan data
print("\nData nilai:", nilai)

# Nilai tertinggi & terendah
nilai_tertinggi = max(nilai)
nilai_terendah = min(nilai)

print("Nilai tertinggi:", nilai_tertinggi)
print("Nilai terendah:", nilai_terendah)

# Rata-rata
rata_rata = sum(nilai) / len(nilai)
print("Rata-rata:", rata_rata)

# Jumlah mahasiswa lulus (>= 60)
lulus = 0
for n in nilai:
    if n >= 60:
        lulus += 1

print("Jumlah mahasiswa lulus:", lulus)
print("Jumlah mahasiswa tidak lulus:", len(nilai) - lulus)
```

```python
import matplotlib.pyplot as plt

kategori = ['Tertinggi', 'Terendah']
nilai_grafik = [nilai_tertinggi, nilai_terendah]

plt.figure()
plt.bar(kategori, nilai_grafik)
plt.title("Perbandingan Nilai Tertinggi & Terendah")
plt.xlabel("Kategori")
plt.ylabel("Nilai")
plt.show()
```

```python
labels = ['Lulus', 'Tidak Lulus']
jumlah = [lulus, len(nilai) - lulus]

plt.figure()
plt.pie(jumlah, labels=labels, autopct='%1.1f%%')
plt.title("Persentase Kelulusan Mahasiswa")
plt.show()
```
